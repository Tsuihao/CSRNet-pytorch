```python
from moviepy.editor import VideoFileClip
from IPython.display import HTML
```

```python
yolo = YOLO()
```

```python
def detect_img_workaround(image):

    img = Image.fromarray(image)
    r_image = yolo.detect_image(img)
    result = np.asarray(r_image)
    return result
```

```python
IN_VIDEO='/dataset/hk.mp4'
OUT_VIDEO='/dataset/hk_yolov3.mp4'
clip = VideoFileClip(IN_VIDEO)
out_clip = clip.fl_image(detect_img_workaround) 
%time out_clip.write_videofile(OUT_VIDEO, audio=False)
```
